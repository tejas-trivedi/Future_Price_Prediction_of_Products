# PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_df
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]
```

```python
avocado_prophet_df
```

```python
avocado_prophet_df = avocado_prophet_df.rename(columns = {'Date': 'ds' , 'AveragePrice': 'y'})
```

```python
avocado_prophet_df
```

### Big advantage of Prophet is that there is no need to perform data processing. Just column names need to be changed according to Prophet. Prophet also works with missing data with several outliners.

# DEVELOPING MODEL AND MAKING PREDICTIONS 

```python
m = Prophet()
m.fit(avocado_prophet_df)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure2 = m.plot_components(forecast)
```
